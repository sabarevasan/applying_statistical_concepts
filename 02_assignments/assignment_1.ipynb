### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
Boston
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))

medv: median value of owner-occupied homes in $1000s.

dis: weighted mean of distances to five Boston employment centres.

_(ii)_ What concept â€˜definesâ€™ the plotted line?

Simple Linear Regression

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?

_(iv)_ Now, letâ€™s fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

```python
# (iii)
# Null hypothesis: There is no liner relation ship between medv and dis, yet the model results in a linear plot. Increase in dis does not increase medv.
# Alternative hypothesis: The regression model provides a good initial benchmark, given that more owner-occupied homes have less value while closer to employment centers and the value increases with more distance from the employment centers. Thus, provides a better representation of the data compared to scatter plot.

# (iv) Extract the data and fit the model
Y = Boston['medv']
X = sm.add_constant(Boston['dis'])
model = sm.OLS(Y, X) # to specify the model
results = model.fit() # to fit the model

# Summarize results
summarize(results)
```

Review your model output to answer the following questions (Hint: use the `summary` and `conf_int` functions):    
_(v)_ What are the _coefficient estimates_ for $B_0$ (intercept) and $B_1$ (slope)?

_(vi)_ What are the _standard errors_ for $B_0$ and $B_1$?

_(vii)_ What are the _confidence intervals_ for $B_0$ and $B_1$?  


```python
# (v) Coefficient estimates for $B_0$ = 18.3901 and $B_1$ = 1.0916

# (vi) Standard errors for $B_0$ = 0.817 and $B_1$ = 0.188

# (vii) Confidence intervals as follows:

# Confidence Intervals
print(results.conf_int())
```


Now, let's interpret the model output.  
_(viii)_ Is the model a good fit? (Hint: review $R^2$)  
_(ix)_ Do we reject the (implied) null hypothesis? Why or why not? (Hint: review model $F$ statistic, $p$ value).  

```python
print(results.summary())

# (viii) From the r-squared (range from 0 to 1) we can see that very little of the variation in medv is explained by dis. It is 0.062 (closer to zero) which means model does not explain much. This is an indication that there are other variables in the data set that have might have an influence on the model.

# (ix) F-statistic signifies the overall effectiveness of the model. At 33.58 the regression model provides a better fit for the dataset than that with no predictors. Yet the p-value is very low indicating that it is unlikely the observed relationship happened by chance, suggesting that th emodel is significant.
```

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
Y = Boston['medv']
X = MS(['dis', 'rm']).fit_transform(Boston)
X = sm.add_constant(X)
model = sm.OLS(Y, X) # to specify the model
results = model.fit() # to fit the model
summarize(results)
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

```python
print(results.summary())
# (ii)
# Null hypothesis: No relationship between the 'medv' and the predictors 'dis' and 'rm'. The F-statistic, and the low p-value, indicates that the model is statistically significant.
# Alternative hypothesis: There is significant relationship between 'medv' and at least one of the predictors 'dis' and 'rm' and is true.

# (iii)
# R-squared: 0.496 - This implies that 49.6% of variance in 'medv' value is explained by using 'dis' and 'rm' predictors which is much greater than that represented by 'dis' variable.
# F-statistic: 247.0 Prob (F-statistic): 1.84e-75 - The value of F-statistic is high and P-value of F-statistic is low. These indicate both 'dis' and 'rm' are good predictors of 'medv'

# (iv)
fig, axes = plt.subplots(figsize=(12, 6))
sm.graphics.plot_regress_exog(results, 'dis', fig=fig)
plt.show()

fig, axes = plt.subplots(figsize=(12, 6))
sm.graphics.plot_regress_exog(results, 'rm', fig=fig)
plt.show()

# Outliers are data points that deviate significantly from the main cluster. In partial regression plots, outliers for dis are extreme values like ~40 and below -20; for rm, they are values over 30 and below -15. These outliers can distort regression coefficients, affecting model fit, leading to incorrect inferences and unreliable predictions.
```

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

```python
# Interaction between 'dis' (distance) and 'rm' (number of rooms) indicates that the influence of distance on the median housing value ('medv') depends on the number of rooms. Specifically, the effect of 'dis' on 'medv' could be more significant for houses with more rooms compared to those with fewer rooms, suggesting that room count moderates the relationship between distance and housing value.

Y = Boston['medv']
X = MS(['dis', 'rm']).fit_transform(Boston)

# interaction term
X["dis_rm"] = X["dis"] * X["rm"] 
X = sm.add_constant(X)

model = sm.OLS(Y, X) # to specify the model
results = model.fit() # to fit the model

print(results.summary())
```
